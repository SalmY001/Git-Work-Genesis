```python
table_bs = BeautifulSoup(table, 'html5lib')
```

# find All

```python
# The find_all() method looks through a tag’s descendants and retrieves all descendants that match your filters.

# The Method signature for find_all(name, attrs, recursive, string, limit, **kwargs)
```

# Name

```python
table_rows=table_bs.find_all('tr')
table_rows
```

```python
first_row =table_rows[0]
first_row
```

```python
print(type(first_row))
```

```python
# Obtain the child
```

```python
first_row.td
```

```python
for i,row in enumerate(table_rows):
    print("row",i,"is",row)
```

```python
# As row is a cell object, we can apply the method find_all to it and extract table cells in the object cells using the tag td, this is all the children with the name td. The result is a list, each element corresponds to a cell and is a Tag object, we can iterate through this list as well.
```

```python
for i,row in enumerate(table_rows):
    print("row",i)
    cells=row.find_all('td')
    for j,cell in enumerate(cells):
        print('colunm',j,"cell",cell)
```

```python
list_input=table_bs .find_all(name=["tr", "td"])
list_input
```
