# Language translator

```python
from ibm_watson import LanguageTranslatorV3
```

```python
url_lt='https://api.eu-gb.language-translator.watson.cloud.ibm.com/instances/16830417-8d53-40f5-a2dd-f75e35e88389'
```

```python
apikey_lt='cZ1EgXnlpdUqN_tyjPSDK5MMmba9SvNo052piJ6Zxe3B'
```

# API requests require a version parameter that takes a date in the format version=YYYY-MM-DD

```python
version_lt='2018-05-01'
```

# create a Language Translator object language_translator:

```python
authenticator = IAMAuthenticator(apikey_lt)
language_translator = LanguageTranslatorV3(version=version_lt,authenticator=authenticator)
language_translator.set_service_url(url_lt)
language_translator
```


# Lists the languages that the service can identify. The method Returns the language code.

```python
from pandas import json_normalize

json_normalize(language_translator.list_identifiable_languages().get_result(), "languages")
```

# We can use the method translate this will translate the text. The parameter text is the text. Model_id is the type of model we would like to use use we use list the language . In this case, we set it to 'en-es' or English to Spanish

```python
translation_response = language_translator.translate(\
    text=recognized_text, model_id='en-es')
translation_response
```

# The result is a dictionary.

```python
translation=translation_response.get_result()
translation
```

# We can obtain the actual translation as a string as follows:

```python
spanish_translation =translation['translations'][0]['translation']
spanish_translation
```

# We can translate back to English

```python
translation_new = language_translator.translate(text=spanish_translation ,model_id='es-en').get_result()
```

```python
translation_eng=translation_new['translations'][0]['translation']
translation_eng
```
